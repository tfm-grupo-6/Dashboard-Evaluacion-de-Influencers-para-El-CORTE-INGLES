### Obtencion Demográficas

Para la extracción de los datos demográficos (datos_demograficos), nos centramos en el contenido ubicado dentro de una etiqueta article que contiene un subtítulo h4 con el texto "Age & Gender". A partir de esta estructura, navegamos por los elementos div internos para identificar los grupos de edad y los porcentajes asociados a mujeres y hombres.

Durante la extracción, observamos que algunos porcentajes estaban duplicados, probablemente debido a alguna inconsistencia en el HTML. Para solventarlo y poder recuperar correctamente los datos, decidimos dividir las repeticiones por la mitad, conservando solo una copia por cada par duplicado. 

```python
def datos_demograficos(perfil):
    ruta_html = f"{perfil}_Modash.io.html"   
    
    with open(ruta_html, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "html.parser")

    divs = []
    for articulo in soup.find_all("article"):
        h4 = articulo.find("h4")
        if h4 and "age & gender" in h4.get_text(strip=True).lower():
            divs = articulo.find_all("div")
            break

    # Extraer porcentajes
    all_percentages = []
    for div in divs:
        texto = div.get_text(strip=True)
        if "%" in texto and re.match(r"\d{1,3}\.\d{2}%", texto):
            all_percentages.append(texto)

    # Contar y conservar solo la mitad de cada repetición
    counter = Counter(all_percentages)
    porcentajes_filtrados = []
    for valor, cantidad in counter.items():
        repeticiones = cantidad // 2  
        porcentajes_filtrados.extend([valor] * repeticiones)

    if len(porcentajes_filtrados) < 10:
        raise ValueError(f"No hay suficientes porcentajes válidos en {perfil} después de filtrar.")

    porcentajes = porcentajes_filtrados[:10]  # solo los primeros 10

    datos = {
        "Mujer_45-64_%": porcentajes[0],
        "Hombre_45-64_%": porcentajes[1],
        "Mujer_35-44_%": porcentajes[2],
        "Hombre_35-44_%": porcentajes[3],
        "Mujer_25-34_%": porcentajes[4],
        "Hombre_25-34_%": porcentajes[5],
        "Mujer_18-24_%": porcentajes[6],
        "Hombre_18-24_%": porcentajes[7],
        "Mujer_13-17_%": porcentajes[8],
        "Hombre_13-17_%": porcentajes[9],
    }

    return datos
```

### Obtencion Geograficas

Para la parte geográfica (extraer_geografica), accedimos de manaera similar a las demograficas al "h4", pero con el texto "Countries". Dentro de este bloque, los nombres de país y sus respectivos porcentajes aparecen "divs" dentro de la clases tipo "flex items-center justify-between". En este caso, filtramos por el texto "Spain" y extraemos el valor del siguiente "div".

```python
from bs4 import BeautifulSoup

def datos_geografica(perfil):
    ruta_html = f"{perfil}_Modash.io.html"  
    
    with open(ruta_html, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "html.parser")

    # Buscar sección "Countries"
    for articulo in soup.find_all("article"):
        h4 = articulo.find("h4")
        if h4 and "countries" in h4.get_text(strip=True).lower():
            divs_contenedor = articulo.find_all("div", class_="flex items-center justify-between")
            for fila in divs_contenedor:
                divs = fila.find_all("div")
                if len(divs) == 2:
                    pais = divs[0].get_text(strip=True).lower()
                    porcentaje = divs[1].get_text(strip=True)
                    if pais == "spain":
                        return {"Geografica_%": porcentaje}

    # Valor por defecto si no se encuentra Spain
    return {"Geografica_%": "0.00%"}
```

```python
# Generar df
resultados = []

for perfil in lista_perfiles:
    datos = {"Perfil": perfil}
    datos.update(datos_demograficos(perfil))
    datos.update(datos_geografica(perfil))
    resultados.append(datos)

# Crear DataFrame final
df_demografica_y_geografica = pd.DataFrame(resultados)

# Transformar a numero los valores
for colum in df_demografica_y_geografica.columns:
    if colum != "Perfil":  # o cualquier otra columna que quieras excluir
        df_demografica_y_geografica[colum] = df_demografica_y_geografica[colum].str.replace('%', '', regex=False).astype(float)
```

### Obtención Links Publicaciones

Detectamos que todas las URLs de publicaciones en Instagram contienen el patrón “/p/”, lo que nos permitió filtrarlas fácilmente.
Una de las dificultades que encontramos fue que, para poder extraer todas las publicaciones visibles en el perfil, era necesario realizar un scroll progresivo en la página. Establecimos un límite de 60 desplazamientos, ya que en cada uno Instagram carga 9 nuevas publicaciones. De este modo, nos aseguramos de cubrir hasta 540 publicaciones por perfil, lo que garantiza, en la mayoría de los casos, abarcar al menos los últimos 12 meses de actividad, incluso para usuarios muy activos.

```python
def obtener_urls_publicaciones(perfil):
    options = Options()
    #options.add_argument("--headless")  # si quieres modo visible, comenta esta línea
    options.add_argument("--user-data-dir=C:/Users/carlos/AppData/Local/Google/Chrome/User Data")  # Ruta a tu perfil
    options.add_argument("--profile-directory=Default")  # O el nombre del perfil, por ejemplo: "Profile 1"
    options.add_argument("--window-size=1920,1080")
    options.binary_location = r"C:/Users/carlos/anaconda_projects/chrome-win64/chrome.exe"  # tu navegador personalizado

    service = Service("C:/Users/carlos/anaconda_projects/chromedriver-win64/chromedriver.exe")
    driver = webdriver.Chrome(service=service, options=options)

    try:
        url = f"https://www.instagram.com/{perfil}"
        driver.get(url)
        time.sleep(20)

        # Aceptar cookies si aparece
        #try:
            #aceptar_btn = driver.find_element(By.XPATH, "//button[contains(text(), 'Permitir todas las cookies')]")
            #aceptar_btn.click()
            #time.sleep(5)
        #except:
            #pass  

        # Scroll dinámico: para obtener más publicaciones
        SCROLL_PAUSE_TIME = 7
        last_height = driver.execute_script("return document.body.scrollHeight")
        urls = set()
        scrolls = 0
        max_scrolls = 60  

        while scrolls < max_scrolls:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(SCROLL_PAUSE_TIME)
            new_height = driver.execute_script("return document.body.scrollHeight")

            # Recoger nuevas URLs tras cada scroll
            soup = BeautifulSoup(driver.page_source, "html.parser")
            for a in soup.find_all("a", href=True):
                href = a["href"]
                if "/p/" in href:
                    urls.add("https://www.instagram.com" + href )

            if new_height == last_height:
                break
            last_height = new_height
            scrolls += 1

    
    finally:
        driver.quit()

    return pd.DataFrame({"Perfil": [perfil] * len(urls), "url": list(urls)})
```

```python
for perfil in lista_perfiles:
    df_urls = obtener_urls_publicaciones(perfil)
    df_publicaciones = pd.concat([df_publicaciones, df_urls]).drop_duplicates()
```

```python
def obtener_urls_reels(perfil):
    options = Options()
    #options.add_argument("--headless")  # si quieres modo visible, comenta esta línea
    options.add_argument("--user-data-dir=C:/Users/carlos/AppData/Local/Google/Chrome/User Data")  # Ruta a tu perfil
    options.add_argument("--profile-directory=Default")  # O el nombre del perfil, por ejemplo: "Profile 1"
    options.add_argument("--window-size=1920,1080")
    options.binary_location = r"C:/Users/carlos/anaconda_projects/chrome-win64/chrome.exe"  # tu navegador personalizado

    service = Service("C:/Users/carlos/anaconda_projects/chromedriver-win64/chromedriver.exe")
    driver = webdriver.Chrome(service=service, options=options)

    try:
        url = f"https://www.instagram.com/{perfil}/reels/"
        driver.get(url)
        time.sleep(20)

        # Aceptar cookies si aparece
        #try:
            #aceptar_btn = driver.find_element(By.XPATH, "//button[contains(text(), 'Permitir todas las cookies')]")
            #aceptar_btn.click()
            #time.sleep(5)
        #except:
            #pass  

        # Scroll dinámico: para obtener más publicaciones
        SCROLL_PAUSE_TIME = 7
        last_height = driver.execute_script("return document.body.scrollHeight")
        urls = set()
        scrolls = 0
        max_scrolls = 45  # puedes subir esto si quieres más profundidad

        while scrolls < max_scrolls:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(SCROLL_PAUSE_TIME)
            new_height = driver.execute_script("return document.body.scrollHeight")

            # Recoger nuevas URLs tras cada scroll
            soup = BeautifulSoup(driver.page_source, "html.parser")
            for a in soup.find_all("a", href=True):
                href = a["href"]
                if "/reel/" in href:
                    urls.add("https://www.instagram.com" + href )

            if new_height == last_height:
                break
            last_height = new_height
            scrolls += 1

    
    finally:
        driver.quit()

    return pd.DataFrame({"Perfil": [perfil] * len(urls), "url": list(urls)})
```

### Obtención Alcance

Para la correcta definición de las delimitaciones temporales, información fundamental para las métricas de rendimiento global, es necesario transformar los datos de publicaciones obtenidos, concretamente la información referente a la columna fecha, a tipo datetime.

En el caso del alcance, se seleccionará un rango temporal de 12 meses respecto a la fecha de inicio de la obtención de datos, el 15/05/2025. No obstante, dado que las publicaciones más recientes pueden no haber alcanzado aún su alcance real, se limitará la ejecución a aquellas publicadas al menos 7 días antes, es decir, del día 08-05-2024 al 08-05-2025.

Además, como la base de datos incluye publicaciones tipo foto sin visualizaciones y algún link dañado, excluiremos los valores nulos y los iguales a cero en esta columna. Finalmente, agruparemos los datos por perfiles y obtendremos como resultado la media de visualizaciones de cada influencer.

```python
#Crear columna con fecha corregida
df_datos_publicaciones["Fecha_datatime"] = pd.to_datetime(df_datos_publicaciones["fecha"], errors="coerce")
```

```python
# Límite temporal
fecha_inicio = pd.Timestamp("2024-05-08")
fecha_fin = pd.Timestamp("2025-05-08")

# Filtro de publicaciones de video en el rango específico
df_videos_ultimo_anno = df_datos_publicaciones[
    (df_datos_publicaciones["Fecha_datatime"] >= fecha_inicio) &
    (df_datos_publicaciones["Fecha_datatime"] < fecha_fin) &
    (df_datos_publicaciones["visualizaciones"].fillna(0) > 0)
]
```

```python
# Agrupación
df_rendimiento = df_videos_ultimo_anno.groupby("Perfil", as_index=False)["visualizaciones"].mean()

# Redondeo
df_rendimiento["visualizaciones"] = df_rendimiento["visualizaciones"].round(0)

# Cambio nombre columna
df_rendimiento.rename(columns={"visualizaciones": "Alcance"}, inplace=True)
```

### Obtención Crecimiento

Para la obtención de este dato, realizaremos un nuevo filtro temporal en los datos de publicaciones, pero esta vez solo para el último mes, de modo que la fecha límite será el 15-04-2025. Después, calcularemos el porcentaje respecto al alcance.

En el caso de que los perfiles no hayan subido contenido en formato vídeo durante los últimos 30 días, se les asignará un valor de 0 en las métricas correspondientes, reflejando así la ausencia de actividad reciente en este tipo de formato.

```python
# Límite temporal
fecha_inicio = pd.Timestamp("2025-04-08")
fecha_fin = pd.Timestamp("2025-05-08")

# Filtro de publicaciones de video en el rango específico
df_videos_ultimo_mes = df_datos_publicaciones[
    (df_datos_publicaciones["Fecha_datatime"] >= fecha_inicio) &
    (df_datos_publicaciones["Fecha_datatime"] < fecha_fin) &
    (df_datos_publicaciones["visualizaciones"].fillna(0) > 0)
]
```

```python
# Agrupación
df_visitas_mes = df_videos_ultimo_mes.groupby("Perfil", as_index=False)["visualizaciones"].mean()

# Redondeo
df_visitas_mes["visualizaciones"] = df_visitas_mes["visualizaciones"].round(0)
```

```python
# Unir ambos DataFrames por la columna 'Perfil'
df_rendimiento = pd.merge(df_rendimiento, df_visitas_mes, on="Perfil", how="left")

# Crear la columna 'Crecimiento_%'
df_rendimiento["Crecimiento_%"] = (df_rendimiento["Alcance"] / df_rendimiento["visualizaciones"]) * 100

# Redondear
df_rendimiento["Crecimiento_%"] = df_rendimiento["Crecimiento_%"].round(0)

# Cambiar nan por 0
df_rendimiento.fillna(0, inplace=True)

# Eliminamos la columna visualizaciones
df_rendimiento = df_rendimiento.drop("visualizaciones", axis=1)
```

## Lenguaje y tono ético 

Para determinar el lenguaje y el tono ético de un influencer, se utilizará el DataFrame df_datos_publicaciones, el cual contiene las descripciones de sus publicaciones. Estas serán analizadas mediante un modelo preentrenado de inteligencia artificial, capaz de identificar contenido tóxico o inapropiado.

```python
# Modelo
classifier = pipeline("text-classification", model="textdetox/xlmr-large-toxicity-classifier")
```

### Obtención Lenguaje y tono ético

A partir de este análisis, se generará una nueva base de datos enriquecida con los resultados del modelo, que incluirá las etiquetas de clasificación (“toxic” o “neutral”) y un score de confianza asociado a cada predicción.

Con el fin de evitar errores, ejecuciones innecesarias y posibles bloqueos por parte de la API, las publicaciones que no dispongan de descripción serán excluidas del análisis automáticamente.

```python
def obtener_lenguaje(fila):
    perfil = fila["Perfil"]
    texto = fila["descripcion"]

    if pd.isna(texto) or str(texto).strip() == "":
        label = None
        score = None
    else:
        resultado = classifier(str(texto))[0]
        label = resultado["label"]
        score = resultado["score"]

    return {"Perfil": perfil, "descripcion": texto, "label": label, "score": score}
```

```python
# Crear DataFrame vacío
df_lenguaje = pd.DataFrame()
```

```python
for i in range(22200,len(df_datos_publicaciones)):
    if i % 100 == 0 and i > 0:
        print(f"⏸ Pausa de 1 minuto tras {i} publicaciones...")
        time.sleep(60)

    try:
        fila = df_datos_publicaciones.iloc[i]
        resultado = obtener_lenguaje(fila)

        # Añadir a resultados y guardar
        df_lenguaje = pd.concat([df_lenguaje, pd.DataFrame([resultado])], ignore_index=True).drop_duplicates()
        df_lenguaje.to_csv("df_lenguaje.csv", index=False, encoding="utf-8")

    except Exception as e:
        print(f" Error en fila {i}: {e}")
```

A continuación, se realizará un conteo por influencer. Si un perfil acumula cinco o más publicaciones que cumplan con estos criterios, será considerado no apto desde el punto de vista comunicativo (valorado como False). En caso contrario, se considerará que mantiene un lenguaje adecuado (valorado como True).

```python
# Descripciones no eticas
df_toxicos = df_lenguaje[(df_lenguaje['label'] == 'toxic') & (df_lenguaje['score'] >= 0.9993)]
```
